## Step 3: Standardize the Features

```python
from sklearn.preprocessing import StandardScaler

# Separate features and standardize
X = df.iloc[:, 1:]  # excluding the class label (assuming it's the first column)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# After standardization, all features had mean - 0 and standard deviation - 1.
```

## Step 4: Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

pca = PCA()  # create PCA object
X_pca = pca.fit_transform(X_scaled)  # fit PCA on your scaled data
```

```python
# This transforms the original 13 features into principal components (PCs) that capture the max variance.
```

## Step 5: Determine Optimal Number of PCs

```python
import numpy as np
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
print("explained_variance:",explained_variance) 
print("cumulative_variance:", cumulative_variance)
```

```python
# First 3 PCs captured ~66% of the variance.
# So, 3 components were selected for further analysis to reduce dimensionality while retaining major information.
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by Principal Components")
plt.grid(True)
plt.show()
```

## Step 6: Apply Clustering on Original Data

```python
from sklearn.cluster import KMeans
kmeans_original = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans_original.fit_predict(X_scaled)
print("clusters_original:",clusters_original)
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score
score_original = silhouette_score(X_scaled, clusters_original)
print(f"Silhouette Score : {round(score_original, 4)}")
```

```python
# KMeans was applied to the original features (before PCA).
# Evaluated using Silhouette Score, which was relatively low (0.28) â€” indicating overlapping or unclear clusters.
```
